```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.linspace(-10, 10, 100) * 0.1
```

```python
y = X ** 3 + 9 * X ** 2 + 9 * X
```

```python
y_p = 3 * X ** 2 + 18 * X + 9
```

```python
plt.plot(X, y)
```

```python
learning_rate = 0.005
```

```python
n = 1
n_list = []
prev_1st = 0
prev_2nd = 0
# while n >= -10 and n <= 10:
while n >= -1 and n <= 1:
    if 3 * n ** 2 + 18 * n + 9 > 0:
        n -= learning_rate * (3 * n ** 2 + 18 * n + 9)
        if n == prev_2nd:
            break
        prev_2nd = prev_1st
        prev_1st = n
        n_list.append(n)
    elif 3 * n ** 2 + 18 * n + 9 < 0:
        n += learning_rate
        if n == prev_2nd:
            break
        prev_2nd = prev_1st
        prev_1st = n
        n_list.append(n)
    else:
        n_list.append(n)
        break
    print(n)
```

```python
plt.plot(X, y)
plt.scatter(n, n ** 3 + 9 * n ** 2 + 9 * n, c = 'r')
```

```python
plt.plot(X, y)
n_list = np.array(n_list)
plt.scatter(n_list[:-1], n_list[:-1] ** 3 + 9 * n_list[:-1] ** 2 + 9 * n_list[:-1], c = n_list[:-1])
plt.scatter(n, n ** 3 + 9 * n ** 2 + 9 * n, c = 'red')
plt.colorbar(ax = plt.gca())
plt.title('Gradient Descent')
plt.show()
```
